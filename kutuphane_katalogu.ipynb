```
import sys
from google.colab import drive
drive.mount("/gdrive",force_remount=True)


dosya_yolu = "/gdrive/My Drive/veritabani.txt"

eserler = []
def oku_yaz(s = str,yazilacak = ""):
    eser_listesi = []
    f = None
    if s == "a":
        f = open(dosya_yolu,"a",encoding="cp1254")
        f.write(yazilacak)
        f.close()
    elif s == "r":
        f = open(dosya_yolu,"r",encoding = "cp1254")
        for i in f.readlines():
            eser_listesi.append([j.strip() for j in i.split('|')])
        f.close()
        return eser_listesi
    elif s == "w":
        f = open(dosya_yolu,"w",encoding = "cp1254")
        f.write(yazilacak)
        f.close()
    else:
        return None


def ara(arama_tipi = str,aranacak_kelime = str):
    bul = None
    for i in eserler:
        if arama_tipi == "y":
            if i[0] == aranacak_kelime:
                 bul = i
                 break
        elif arama_tipi == "e":
            if i[1] == aranacak_kelime:
                bul = i
                break

        elif arama_tipi == "b":
            if i[2] == aranacak_kelime:
                bul = i
                break
        elif arama_tipi == "a":
            if i[3] == aranacak_kelime:
                bul = i
                break

    return bul

def katalog():
    global eserler
    eserler = oku_yaz("r")

    print("Hoşgeldiniz")
    while 1:
        c = input("E-> Eser ekleme\nA-> Arama\nD-> Eser silme\nQ-> Tüm eserleri göster\nW-> Programdan çıkış\nSeçim Yapınız..: ")

        if c == "E":
            yeni_eser = [input("Eser adı giriniz: "),input("Yazar ismi giriniz: "),input("Basım yılı giriniz: "),input("Yayınevi giriniz: ")]
            eserler.append(yeni_eser)
            yazdir = "|".join(yeni_eser)
            oku_yaz("a",yazdir.strip())
            print("Eser eklendi...")
        elif c == "A":
            arama_tipi = input("y-> Yazar araması\ne-> Eser araması\nb-> Basım yılı araması\na-> Yayınevi araması\nSeçim yapınız..: ")
            aranacak_kelime = input("Aranacak kelimeyi giriniz: ")

            bul = ara(arama_tipi,aranacak_kelime)
            if bul == None:
                print("Sonuç bulunamadı. ")
            else:
                print(bul[1]+"\n",bul[0]+"\n"+bul[2]+"\n"+bul[3]+"\n\n")
        elif c == "D":
            silinecek = input("Silinecek eser adı giriniz: ")
            for i in eserler:
                if i[1] == silinecek:
                    eserler.remove(i)
            yeni = ""
            for j in eserler:
                yeni += j[1]+","+j[0]+","+j[2]+","+j[3]
            oku_yaz("w",yeni)
            print("Eser silindi. ")
        elif c == "Q":
            for i in eserler:
                print("\nEser adı: "+i[1]+"\nEser yazarı: "+i[0]+"\nEser basım yılı: "+i[2]+"\nEser yayınevi: "+i[3]+"\n")
        elif c == "W":
            print("Çıkış yapıldı. ")
            break
        else:
            print("Yanlış bir seçim yaptınız. ")

katalog()
```
